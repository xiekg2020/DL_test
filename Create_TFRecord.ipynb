```python
paths = {
    'SCRIPTS_PATH': os.path.join('GenerateTFRecord'),
    'ANNOTATION_PATH': os.path.join('drive','MyDrive','workspace','annotations'),
    'IMAGE_PATH': os.path.join('drive','MyDrive','workspace','images'),
 }
```

```python
TF_RECORD_SCRIPT_NAME = 'generate_tfrecord.py'
LABEL_MAP_NAME = 'label_map.pbtxt'
files = {
    'TF_RECORD_SCRIPT': os.path.join(paths['SCRIPTS_PATH'], TF_RECORD_SCRIPT_NAME), 
    'LABELMAP': os.path.join(paths['ANNOTATION_PATH'], LABEL_MAP_NAME)
}
```

```python
labels = [{'name':'helmet', 'id':1}, {'name':'head', 'id':2}]
with open(files['LABELMAP'], 'w') as f:
    for label in labels:
        f.write('item { \n')
        f.write('\tname:\'{}\'\n'.format(label['name']))
        f.write('\tid:{}\n'.format(label['id']))
        f.write('}\n')
```

```python
! python {files['TF_RECORD_SCRIPT']} -x {os.path.join(paths['IMAGE_PATH'], 'train')} -l {files['LABELMAP']} -o {os.path.join(paths['ANNOTATION_PATH'], 'train.tfrecord')}
```
